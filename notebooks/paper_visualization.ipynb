```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import to_rgb, to_hex

# Define your base color
base_color = '#E0E0E0'  # Light orange

def get_luminance(hex_color):
    """Returns perceived brightness of a color [0–1]"""
    r, g, b = to_rgb(hex_color)
    return 0.299 * r + 0.587 * g + 0.114 * b


# Function to generate shades from a single base color
def generate_shades(base_hex, n):
    base_rgb = np.array(to_rgb(base_hex))
    factors = np.linspace(0.6, 1.0, n)  # lighter to darker shades
    return [to_hex(base_rgb * f) for f in factors]

# Real-world data
real_data = {
    "Attribute": [
        "Axes Orientation", "Axes Orientation",
        "Numerical Range", "Numerical Range", "Numerical Range",
        "Category Colors", "Category Colors",
        "Text Length",
        "Axes Count", "Axes Count", "Axes Count",
        "Data Points", "Data Points", "Data Points",
        "Number of Categories", "Number of Categories", "Number of Categories"
    ],
    "Category": [
        "Vertical", "Horizontal",
        "<1", "1-9", "3-4 digits",
        "Distinct", "Single",
        "1–15 chars",
        "3", "4", "5",
        "70 lines", "30 lines", "20 lines",
        "3", "4", "5"
    ],
    "Percentage": [
        80, 20,
        50, 25, 25,
        60, 40,
        100,
        30, 40, 30,
        60, 30, 10,
        50, 25, 25
    ]
}

# Synthetic data
synthetic_data = {
    "Attribute": [
        "Axes Orientation",
        "Numerical Range", "Numerical Range", "Numerical Range",
        "Category Colors",
        "Text Length",
        "Axes Count", "Axes Count", "Axes Count",
        "Data Points",
        "Number of Categories", "Number of Categories", "Number of Categories"
    ],
    "Category": [
        "Vertical",
        "<1", "1-9", "3-4 digits",
        "Distinct",
        "1–15 chars",
        "3", "4", "5",
        "100 lines",
        "3", "4", "5"
    ],
    "Percentage": [
        100,
        50, 25, 25,
        100,
        100,
        30, 40, 30,
        100,
        50, 25, 25
    ]
}

# Create DataFrames
df_real = pd.DataFrame(real_data)
df_synth = pd.DataFrame(synthetic_data)

# Add source labels
df_real['Source'] = 'Real'
df_synth['Source'] = 'Synthetic'

# Combine and use MultiIndex
df_combined = pd.concat([df_real, df_synth], ignore_index=True)
df_combined.set_index(['Attribute', 'Source'], inplace=True)

# Pivot
pivot = df_combined.pivot_table(index=['Attribute', 'Source'], columns='Category', values='Percentage').fillna(0)

# Plot settings
fig, ax = plt.subplots(figsize=(10, 5))
bar_width = 0.35
spacing = 0.15

attributes = pivot.index.get_level_values(0).unique()
categories = pivot.columns

x_positions = []
x_labels = []

# Loop through attributes
for i, attr in enumerate(attributes):
    real = pivot.loc[(attr, 'Real')]
    synth = pivot.loc[(attr, 'Synthetic')]

    # Filter to only non-zero categories
    all_values = pd.concat([real, synth])
       # Filter to only non-zero categories for this attribute
    present_categories = [cat for cat in categories if (real.get(cat, 0) + synth.get(cat, 0)) > 0]

    # Generate N shades from base color
    shades = generate_shades(base_color, len(present_categories))
    category_colors = dict(zip(sorted(present_categories), shades))  # consistent shade order


    x_real = i * (2 * bar_width + spacing)
    x_synth = x_real + bar_width

    bottom_real = 0
    bottom_synth = 0

    for cat in sorted(present_categories):
        val_r = real.get(cat, 0)
        val_s = synth.get(cat, 0)
        color = category_colors[cat]

        # Real bar segment
        if val_r > 0:
            ax.bar(x_real, val_r, bottom=bottom_real, width=bar_width, color=color, edgecolor='black', linewidth=0.5)
            if val_r > 5:
                # fontsize = 8 if len(cat) <= 10 else 7
                fontsize = 6
                rotation = 90 if len(cat) <= 10 else 0
                text_color = 'white' if get_luminance(color) < 0.5 else 'black'
                ax.text(x_real, bottom_real + val_r / 2, cat,
                        ha='center', va='center', fontsize=fontsize,
                        color=text_color, rotation=rotation)

                # ax.text(x_real, bottom_real + val_r / 2, cat, ha='center', va='center', fontsize=fontsize, color='black', rotation=rotation)
                
            
            if bottom_real > 0:
                ax.plot([x_real - bar_width / 2, x_real + bar_width / 2], [bottom_real, bottom_real], color='black', linewidth=1)
            bottom_real += val_r

        # Synthetic bar segment
        if val_s > 0:
            ax.bar(x_synth, val_s, bottom=bottom_synth, width=bar_width, color=color, edgecolor='black', linewidth=0.5)
            if val_s > 5:
                fontsize = 6
                rotation = 90 if len(cat) <= 10 else 0
                text_color = 'white' if get_luminance(color) < 0.5 else 'black'
                ax.text(x_synth, bottom_synth + val_s / 2, cat, ha='center', va='center', fontsize=fontsize, color=text_color, rotation=rotation)

                # ax.text(x_synth, bottom_synth + val_s / 2, cat, ha='center', va='center', fontsize=fontsize, color='black', rotation=rotation)
            if bottom_synth > 0:
                ax.plot([x_synth - bar_width / 2, x_synth + bar_width / 2], [bottom_synth, bottom_synth], color='black', linewidth=1)
            bottom_synth += val_s

    # R and S labels
    ax.text(x_real, bottom_real + 2, "R", ha='center', va='bottom', fontsize=7)
    ax.text(x_synth, bottom_synth + 2, "S", ha='center', va='bottom', fontsize=7)

    # Vertical divider
    mid = (x_real + x_synth) / 2
    ax.axvline(mid, color='black', linewidth=2)

    # X-tick label
    x_center = (x_real + x_synth) / 2
    x_positions.append(x_center)
    x_labels.append(attr)

# Final touches
ax.set_xticks(x_positions)
ax.set_xticklabels(x_labels, fontsize=7)
ax.set_ylabel("Percentage (%)", fontsize=8)
ax.set_ylim(0, 110)
ax.set_title("Attribute Comparison: Real (R) vs Synthetic (S)", fontsize=9)

plt.tight_layout()
plt.show()
```
